1.이진 트리(Binary Tree)

```python
class TreeNode:
    def __init__(self, data, left, right):
        self.data = data
        self.left = left
        self.right = right 

    def set_left(self, left):
        self.left = left

    def set_right(self, right):
        self.right = right


# 바이너리 트리 생성
d = TreeNode('D', None, None)
e = TreeNode('E', None, None)
b = TreeNode('B', d, e)

f = TreeNode('F', None, None)
c = TreeNode('C', f, None)

a = TreeNode('A', b, c)

my_binary_tree = a
```

```python
# preorder (선위 순회)
def preorder(node):
    # if node == None: return

    print(node.data, end=' ')
    if node.left != None:
        preorder(node.left)
    if node.right != None:
        preorder(node.right)


preorder(my_binary_tree)
```

```python
# inorder (중위 순회)
def inorder(node):
    if node == None:
        return

    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)


inorder(my_binary_tree)
```

```python
# postorder (후위 순회)
def postorder(node):
    if node == None:
        return

    postorder(node.left)
    postorder(node.right)
    print(node.data, end=' ')


postorder(my_binary_tree)
```

```python
# 딕셔너리 트리

tree = {}

tree['A'] = ['B', 'C']
tree['B'] = ['D', 'E']
tree['C'] = ['F', None]
tree['D'] = [None, None]
tree['E'] = [None, None]
tree['F'] = [None, None]
```

```python
# preorder (선위 순회)
def preorder(tree, node):
    if node == None:
        return

    print(node, end=' ')

    preorder(tree, tree[node][0])  # 왼쪽 자식
    preorder(tree, tree[node][1])  # 오른쪽 자식


preorder(tree, 'A')
```

```python
# inorder (중위 순회)
def inorder(tree, node):
    if node == None:
        return

    inorder(tree, tree[node][0])   # 왼쪽 자식
    print(node, end=' ')           # 현재 노드
    inorder(tree, tree[node][1])   # 오른쪽 자식


inorder(tree, 'A')
```

```python
# postorder (후위 순회)
def postorder(tree, node):
    if node == None:
        return

    postorder(tree, tree[node][0])  # 왼쪽 자식
    postorder(tree, tree[node][1])  # 오른쪽 자식
    print(node, end=' ')            # 현재 노드


postorder(tree, 'A')
```

Binary Search Tree

```python
class STreeNode:
    def __init__(self, key, left, right):
        self.key = key
        self.left = left
        self.right = right 
```

```python
# BST 생성
t3 = STreeNode(3, None, None)
t12 = STreeNode(12, None, None)
t7 = STreeNode(7, t3, t12)

t27 = STreeNode(27, None, None)
t31 = STreeNode(31, t27, None)
t26 = STreeNode(26, None, t31)

t18 = STreeNode(18, t7, t26)
my_bst = t18
```

```python
def search_bst(n, key):
    if n == None: return None
    if key == n.key: return n
    if key < n.key:
        return search_bst(n.left, key)
    else:
        return search_bst(n.right, key)

result = search_bst(my_bst, 25)
if result: print(result.key)
else:
    print('없는 key입니다.')
```

```python
def search_bst_iter(n, key):
    while n != None:
        if key == n.key:
            return n
        elif key < n.key:
            n = n.left
        else:
            n = n.right
    return None

result = search_bst_iter(my_bst, 25)
if result: print(result.key)
else:
    print('없는 key입니다.')
```
