```python
def is_prime(n): #함수
    is_prime_number = True

    if n == 1:
        is_prime_number = False
    else:
        for i in range(2, n):
            if n % i == 0:
                is_prime_number = False
                break

    return is_prime_number
    
num = int(input("숫자를 입력하세요: "))

if is_prime(num):
    print("소수입니다.")
else:
    print("소수가 아닙니다.")
```

```python
def is_prime(n):
    if n == 1:
        return False
    elif n == 2:
        return True
    elif n % 2 == 0:
        return False
    else:
        for i in range(3, n // 2 + 1, 2):
            if n % i == 0:
                return False
        return True


num = int(input("숫자를 입력하세요: "))

if is_prime(num):
    print("소수입니다.")
else:
    print("소수가 아닙니다.")
```

```python
def is_prime_candidate(x):
    if x == 1:
        return False
    elif x == 2:
        return True
    elif x % 2 == 0:
        return False
    else:
        for i in range(3, x, 2):
            if x % i == 0:
                return False
        return True


def is_prime(n):
    if n == 1:
        return False
    elif n == 2:
        return True
    elif n % 2 == 0:
        return False
    else:
        for i in range(3, n // 2 + 1, 2):
            if is_prime_candidate(i):   # i가 소수일 때만
                if n % i == 0:
                    return False
        return True


num = int(input("숫자를 입력하세요: "))

if is_prime(num):
    print("소수입니다.")
else:
    print("소수가 아닙니다.")
```

```python
M = int(input("첫 번째 숫자를 입력하세요: "))
N = int(input("두 번째 숫자를 입력하세요: "))

common_divisor_list = []

if M > N:
    M, N = N, M

for num in range(1, M + 1):
    if M % num == 0 and N % num == 0:
        common_divisor_list.append(num)

print(f"공약수: {common_divisor_list}")
```

```python
M = int(input("첫 번째 숫자를 입력하세요: "))
N = int(input("두 번째 숫자를 입력하세요: "))

if M > N:
    M, N = N, M

gcd = 1

for num in range(M, 0, -1):
    if M % num == 0 and N % num == 0:
        gcd = num
        break

print(f"최대공약수: {gcd}")
```

```python
M = int(input("첫 번째 숫자를 입력하세요: "))
N = int(input("두 번째 숫자를 입력하세요: "))

if M > N:
    M, N = N, M

lcm = N
multiple = N

while True:
    if multiple % M == 0:
        lcm = multiple
        break
    multiple += N

print(f"최소공배수: {lcm}")
```
