```python
from collections import deque
queue = deque()

print('시작')

while True:
    # 메뉴 프린트
    # 메뉴 번호 입력
    menu = int(input('>>'))

    if menu == 0: break
    elif menu == 1: #enqueue
        data = int(input('데이터 :'))
        queue.append(data)
    elif menu == 2: # dequeue
        if queue:
            data = queue.popleft()
            print('가져온 데이터 :', data)
        else:
            print('Queue가 비어 있습니다.')
    elif menu == 3: # peek
        if queue:
            print(queue[0])
        else:
            print('Queue가 비어 있습니다.')

    print(queue)
    
print('종료')
```

```python
queue = []


def is_empty():
    return len(queue) == 0


def enqueue(data):
    queue.append(data)


def dequeue():
    if is_empty():
        return None
    return queue.pop(0)


def get_front():
    if is_empty():
        return None
    return queue[0]


print('[[정수형 큐 연산 파이썬 구현 실습]]')

while True:
    print('===================================')
    print('1.Enqueue  2.Dequeue  3.Get Front  0.Exit')
    print('===================================')

    menu = int(input('>> 연산 메뉴 선택 : '))

    if menu == 0:
        print('정수형 Queue 종료')
        break

    elif menu == 1:
        data = int(input('>> 추가할 데이터 입력 : '))
        enqueue(data)
        print('[Queue]', queue)

    elif menu == 2:
        data = dequeue()
        if data is None:
            print('Queue가 비어 있습니다.')
        else:
            print('가져온 데이터 :', data)
        print('[Queue]', queue)

    elif menu == 3:
        data = get_front()
        if data is None:
            print('Queue가 비어 있습니다.')
        else:
            print('Front 데이터 확인 :', data)
        print('[Queue]', queue)

    else:
        print('잘못된 메뉴입니다.')
```
